```python

import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return (1/4)*x**4 - x**2 - x

def f_prime(x):
    return x**3 - 2*x - 1

def forward_difference(f, x, h=0.01):
    return (f(x+h) - f(x))/h

def backward_difference(f, x, h=0.01):
    return (f(x) - f(x-h))/h

def centred_difference(f, x, h=0.01):
    return (f(x+h) - f(x-h))/(2*h)

x = np.linspace(-2, 3, 100)

plt.plot(x,f(x), color='purple',label='f')
plt.plot(x,f_prime(x), color='green',label='f prime')
plt.plot(x, forward_difference(f, x), color='red',label='Forward Difference')
plt.plot(x, backward_difference(f, x), label='Backward Difference')
plt.plot(x, centred_difference(f,x),label='Centered Difference')

plt.legend(loc='upper left')

plt.show()
```

```python


def dudt(t, u):
    x = u[0]
    v = u[1]
    
    k = 1
    
    return [v, -k*x]
   

def runge_kutta1(t0, u0, t, h):
    u = [u0[0], u0[1]]
    n = int((t - t0)/h)
#     t = 0
    
    
    for i in range(1, n + 1):
        
        
        
        
        k1 = [h * dudt(t, u)[0], h * dudt(t, u)[1]]
        
        k2 = [
                h * dudt(t + (h/2), [u[0] +  (k1[0]/2), u[1] +  (k1[1]/2)])[0],
                h * dudt(t + (h/2), [u[0] +  (k1[0]/2), u[1] +  (k1[1]/2)])[1]
        ]
        
        k3 = [
                h * dudt(t + (h/2), [u[0] +  (k2[0]/2), u[1] +  (k2[1]/2)])[0],
                h * dudt(t + (h/2), [u[0] +  (k2[0]/2), u[1] +  (k2[1]/2)])[1]
             ]
        
        k4 = [
              h * dudt(t + h, [u[0] + k3[0], u[1] + k3[1]])[0], 
              h * dudt(t + h, [u[0] + k3[0], u[1] + k3[1]])[1]
             ]
 
  
        u[0] = u[0] + (1/6)*(k1[0] + 2 * k2[0] + 2 * k3[0] + k4[0])
        
        u[1] = u[1] + (1/6)*(k1[1] + 2 * k2[1] + 2 * k3[1] + k4[1])

        t0 = t + h
        
        #print(f"u = {u}")
    return u
 


t0 = 0
u0 = [-6, 0]
t = 7
h = 0.2


 

positions = [runge_kutta1(t0, u0, k, h)[0] for k in range(0, 40)]
theory_positions = [u0[0]*math.cos(k) for k in range(0, 40)] #u(t) = u0*cos(t) (theoretical curve)

ks = [a for  a in range(len(positions))]
print("Runge-Kutta curve")
sns.lineplot(x = ks, y = positions)
plt.show()
print("Theory curve")
sns.lineplot(x = ks, y = theory_positions)
plt.show()

'''This is an undamped oscilator whose amplitude remains constant over time as shown by the curve'''
```
